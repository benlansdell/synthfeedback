```python
for i in range(data['metrics'].shape[0]):
    store_acc=np.mean(data['metrics'][i,:,:,0],axis=0)
    leg="combo%d"%i
    plt.plot(store_acc,label=leg)
plt.legend()
plt.title("loss")
plt.xlabel("Epoch")
plt.show()
```
