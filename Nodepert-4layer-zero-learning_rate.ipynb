# Results for nodepert in a 4 layered network and zero learning rate

Parameters:
batch size=32
Epochs=2000

```python
print(data['tags'])
neuro_num=[30,100,300,1000]
```

```python
data['metrics'].shape
```

```python
for i in range(data['metrics'].shape[0]):
    store_acc=np.mean(data['metrics'][i,:,:,0],axis=0)
    plt.plot(store_acc,label=neuro_num[i])
plt.legend()
plt.title("Acc")
plt.xlabel("Epoch")
plt.show()
```
