```python
import os
os.environ["CUDA_VISIBLE_DEVICES"]='0'

import tensorflow as tf
config = tf.ConfigProto()
config.gpu_options.allow_growth = True
import matplotlib.pyplot as plt
import numpy as np
import pickle
```

```python
with open('/home/prashanth/synthfeedback/experiments/nodepert4_fixedw_exact/summary/2_establish_convergence_feedforward_output.npz','rb') as f:
    data=pickle.load(f,encoding='latin1')
```

```python
num_neuro=[500,1000,1500,2000]
```

```python
data['metrics'].shape
```

# Accuracy

```python
for i in range(data['metrics'].shape[0]):
    store_acc=np.mean(data['metrics'][i,:,:,0],axis=0)
    plt.plot(store_acc,label=num_neuro[i])
plt.legend()
plt.show()
```
