```python
import numpy as np
import csv
import matplotlib.pyplot as plt
```

```python
val1=[]
val2=[]
val3=[]
val4=[]
val5=[]
val6=[]
with open('consolidated.csv') as csvfile:
    readCSV = csv.reader(csvfile, delimiter=',')
    for row in readCSV:
        #print(row)
        val1.append(row[1])
        val2.append(row[2])
        val3.append(row[3])
        val4.append(row[4])
        val5.append(row[5])
        val6.append(row[6])
```

```python
del (val1[0],val2[0],val3[0],val4[0],val5[0],val6[0])
learning_rate=np.logspace(-3,-6,num=6)
learning_rate
```

```python
val1=[float(i) for i in val1]
val2=[float(i) for i in val2]
val3=[float(i) for i in val3]
val4=[float(i) for i in val4]
val5=[float(i) for i in val5]
val6=[float(i) for i in val6]
```

```python
plt.plot(val1,label=learning_rate[0])
plt.plot(val2,label=learning_rate[1])
plt.plot(val3,label=learning_rate[2])
plt.plot(val4,label=learning_rate[3])
plt.plot(val5,label=learning_rate[4])
plt.plot(val6,label=learning_rate[5])
plt.legend(loc='lower right')
plt.show()
```
